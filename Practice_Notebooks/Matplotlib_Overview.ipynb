plot: Plot Lines and show: Display a Figure

```python
x1 = [1, 2, 3, 4] 
y1 = [2, 4, 6, 8]
x2 = [2, 4, 6, 8] 
y2 = [1, 2, 3, 4]
```

```python
plt.plot(x1, y1, label = 'First Line') #label: To assign a name to the line 
plt.plot(x2, y2, label = 'Second Line')
plt.xlabel('X Axis') # xlabel: Assign label to x - axis
plt.ylabel('Y Axis') # ylabel: Assign label to y - axis
plt.title('Simple Graph') # title: The plot’s title 
plt.legend()# legend: Places a legend in the axes 
plt.show()
```

```python
x1 = [1, 2, 3, 4] 
y1 = [2, 4, 6, 8] 
x2 = [2, 4, 6, 8] 
y2 = [1, 2, 3, 4]
plt.plot(x1, y1, color = 'g', linewidth = 3, linestyle = '-', label = 'First Line') 
plt.plot(x2, y2, color = 'r', linewidth = 3, linestyle = '-.', label = 'Second Line')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple Graph')
plt.grid() # grid: Turn the axes grids on or off 
plt.legend()
plt.show()
```

```python
x1 = [1, 2, 3, 4]
y1 = [2, 4, 6, 8] 
x2 = [2, 4, 6, 8]
y2 = [1, 2, 3, 4]
plt.plot(x1, y1, 'g-' , label = 'First Line') # Shortcut Declaration
plt.plot(x2, y2, 'r-.' , label = 'Second Line') #Shortcut Declaration
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple Graph')
plt.grid() #grid: Turn the axes grids on or off 
plt.legend()
plt.show()
```

```python
x1 = [1, 2, 3, 4]
y1 = [2, 4, 6, 8] 
x2 = [2, 4, 6, 8]
y2 = [1, 2, 3, 4]
plt.plot(x1, y1, 'g-', marker = 's', markersize = 10, label = 'First Line') 
plt.plot(x2, y2, 'r-', marker = 'o', markersize = 6, label = 'Second Line')
plt.xlabel('X Axis') 
plt.ylabel('Y Axis') 
plt.title('Simple Graph') 
plt.grid()
plt.legend() 
plt.show()
```

```python
x1 = [1, 2, 3, 4] 
y1 = [2, 4, 6, 8] 
x2 = [2, 4, 6, 8]
y2 = [1, 2, 3, 4]
plt.plot(x1, y1, 'g-*', markersize = 6, label = 'First Line') #Shortcut Declration
plt.plot(x2, y2, 'r-o', markersize = 6, label = 'Second Line') #Shortcut Declration
plt.xlabel('X Axis') 
plt.ylabel('Y Axis') 
plt.title('Simple Graph') 
plt.grid()
plt.legend() 
plt.show()
```

```python
x = np.arange(0, 10,0.2) 
s = np.sin(x)
c = np.cos(x)
plt.plot(x, s, label='Sine Wave') 
plt.plot(x, c, label='Cosine Wave')
plt.xlabel('X Axis') 
plt.ylabel('Y Axis') 
plt.title('Simple Graph') 
plt.grid()
plt.legend() 
plt.show()
```

```python
company = ['Infosys', 'TCS', 'HCL', 'Wipro', 'Google', 'FB', 'Amazon', 'Twitter']
df = pd.DataFrame({'Revenue (in $million)': [200, 240, 300, 140, 600, 770, 250, 450]}, index=company)
df
```

```python
y = df.values 
y
```

```python
y.flatten()
```

```python
y = y.flatten()
xpos = np.arange(8)
plt.bar(xpos, y, label = 'Revenue')
plt.xticks(xpos, company) # xticks: Setting tick labels on the x-axis 
plt.ylabel('Company Revenue in $million')
plt.xlabel('Companies')
plt.title('Bar Chart')
plt.legend()
plt.show()
```

```python
y = df.values 
y = y.flatten()
xpos = np.arange(8)
plt.barh(xpos, y, label = 'Revenue')
plt.yticks(xpos, company) # yticks: Setting tick labels on the y-axis
plt.xlabel('Company Revenue in $million')
plt.title('Bar Chart')
plt.legend()
plt.show()
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth = 0.8, bins = 4) #bins: Number of Bars, default = 10
plt.show()
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129] 
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g')
plt.xlabel("Sugar Level") 
plt.ylabel("Number Of Patients") 
plt.title("Blood Sugar Chart") 
plt.show()
```

Bin Range:
● 80-100: Normal
● 100-125: Pre-Diabetic
● 125-150: Diabetic

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'], label=['Men','Women'])
plt.xlabel("Sugar Level") 
plt.ylabel("Number Of Patients") 
plt.title("Blood Sugar Chart") 
plt.legend()
fig = plt.gcf()
fig.set_size_inches(10,9)
plt.show()
```

```python
expense_values = [1400,600,300,410,250]
expense_labels = ["HomeRent", "Food", "Phone/Internet Bill", "Car Maintenance", "Other Utilities"]
plt.pie(expense_values, labels=expense_labels,radius=2) # pie: Plot a Pie Chart → labels: A sequence of strings providing the labels for each wage
plt.title('Pie Chart') 
#plt.savefig('Pie Chart.jpg')
plt.show()
```

```python
expense_values = [1400,600,300,410,250]
expense_labels = ["HomeRent", "Food", "Phone/Internet Bill", "Car Maintenance", "Other Utilities"]
plt.pie(expense_values, labels=expense_labels)
plt.title('Pie Chart')
plt.axis('equal')# axis: Convenience method to set or get axis properties, and ‘equal’ means x-axis and y-axis have same length i.e. a Circle is Circular
plt.show()
```

```python
expense_values = [1400,600,300,410,250]
expense_labels = ["HomeRent", "Food", "Phone/Internet Bill", "Car Maintenance", "Other Utilities"]
plt.pie(expense_values, labels=expense_labels, autopct='%1.2f%%', radius=1.5) # autopct: To label the wages with their Numeric Values
#radius: The radius of the Pie
plt.title('Pie Chart') 
plt.axis('equal') 
plt.show()
```

```python
expense_values = [1400,600,300,410,250]
expense_labels = ["HomeRent", "Food", "Phone/Internet Bill", "Car Maintenance", "Other Utilities"]
plt.pie(expense_values, labels=expense_labels, autopct='%1.1f%%', radius=1.5, explode=[0,0,0,0,0.2])
# explode: To offset each Wage by specified fraction
plt.title('Pie Chart') 
plt.axis('equal') 
plt.show()
```
